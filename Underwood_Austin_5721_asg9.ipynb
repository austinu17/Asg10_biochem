## Q17


<b>Part A.</b>
<p>TH•A + H*OH* -----> TH* •A + HOH•
<p>(TH•A)<sub>close</sub> <---<sup>k<sub>op</sub></sup> <sub>k<sub>cl</sub></sub> ---> (TH•••A)<sub>open</sub>
<p>(TH•••A)<sub>open</sub> + B + H*OH* --<sup>K<sub>tr</sub></sup>--> (TH*•A)<sub>close</sub>+ B + HOH*
<p>d[(TH*•A)]/dt = K<sub>ex</sub> [(TH•A)<sub>close</sub>]
<p>d[(TH•••A)<sub>open</sub>]/dt = k<sub>op</sub> [(TH•••A)<sub>close</sub>]-k<sub>cl</sub> [(TH•••A)<sub>open</sub>]k<sub>op</sub> [(TH•••)<sub>open</sub>][B] = 0
<p>K<sub>ex</sub>[(TH•A)<sub>closed</sub>]= (k<sub>tr</sub>*k<sub>op</sub>[B][(TH•A)<sub>closeed</sub>])/(k<sub>cl</sub>*k<sub>tr</sub>[B])
<p>K<sub>exp</sub> = (k<sub>tr</sub> * k<sub>op</sub>[B])/(k<sub>cl</sub>+k<sub>tr</sub>[B])
<p><b> Part B.</b>
<p>1/K<sub>exp</sub> = (k<sub>cl</sub>+k<sub>tr</sub>[B])/(k<sub>tr</sub> * k<sub>op</sub>[B])
<p>1/K<sub>exp</sub> = (k<sub>cl</sub>)/(k<sub>tr</sub> * k<sub>op</sub>) * {1/[B]} +1 /K<sub>op</sub>
<p> This allows for the formula to be able to applied to the linear formula of y = mx+B where the intercpet of the formula = 1/k<sub>op</sub>

```python
import numpy as np
import matplotlib.pyplot as plt
print("part c.")
OH = [1.26,2.00,3.16,10,31.6]
K = [12.5,18.2,27.2,69.9,140.8]
K = np.array(K)
K = 1/K
OH = np.array(OH)
OH = 1/OH
plt.plot(OH,K)
plt.ylabel('1/K_ex (1/s)')
plt.xlabel('1/[OH-] (mico_M)')
p_1 = np.polyfit(OH,K,1)
intercept = float(p_1[1])
K_op = 1/intercept
K_op=(float('%.3g'% (float(K_op))))
print("The K_op of the dataset was found to be "+str(K_op)+ " 1/s")
plt.show()
```
